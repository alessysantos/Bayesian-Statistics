```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.naive_bayes import CategoricalNB

#Dataset
data = {
    'Long':   ['Yes']*400 + ['No']*100 + ['Yes']*0   + ['No']*300 + ['Yes']*100 + ['No']*100,
    'Sweet':  ['Yes']*350 + ['No']*150 + ['Yes']*150 + ['No']*150 + ['Yes']*100 + ['No']*100,
    'Yellow': ['Yes']*450 + ['No']*50  + ['Yes']*300 + ['No']*0   + ['Yes']*50  + ['No']*150,
    'Fruit':  ['Banana']*500           + ['Orange']*300           + ['Other']*200
}

df = pd.DataFrame(data)

#Preprocessing
feature_encoder = OrdinalEncoder()
X = feature_encoder.fit_transform(df[['Long', 'Sweet', 'Yellow']])
y = df['Fruit']

#Training the Model
model = CategoricalNB(alpha=1.0)
model.fit(X, y)

#Simulation
test_case = pd.DataFrame([['Yes', 'Yes', 'Yes']], # Corrected the 'Sweet' entry here to 'Yes' for consistency in the test case
                         columns=['Long', 'Sweet', 'Yellow'])

test_encoded = feature_encoder.transform(test_case) # Changed to use test_case DataFrame

#Output Probabilities
probs = model.predict_proba(test_encoded)
prediction = model.predict(test_encoded)

print("--- Fruit Classification Results (ML Plus Reference) ---")
print(f"Test Input: Long=Yes, Sweet=Yes, Yellow=Yes")
for i, fruit_class in enumerate(model.classes_):
    print(f"Probability [{fruit_class}]: {probs[0][i]:.4f}")

print(f"\nFinal Prediction: The fruit is likely a {prediction[0]}")
```
